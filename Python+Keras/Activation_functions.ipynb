# Funciones de activacion

Una red neuronal introduce una nolinealidad a lis datos resultantes de una red neuronal
Esto le permite aprender

## Transforma la salida de cada neurona
Luego de que la neurona realice la suma ponderada de los pesos involucrados y el sesgo, la fucnion de activacion
deside que dato va a pasar a la siguiente neurona

## Introduce no linealidad
Una neurona realiza combinaciones lineales de la informacion que resibe mas los pesos y el sesgo, solo con esta informacion no puede aprender
ni seguir una tendencia

Sin funciones de activacion una red neuronal no puede modelas o aprender mas alla de relaciones que sean de tipo lineal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funcion sigmoide
def sigmoid(a):
    return 1/(1+np.exp(-a))
```

```python
# Funcion escalonada
def step(a):
    return np.piecewise(a, [a <= 0.0, a > 0.0], [0, 1])
```

```python
# Funcion ReLu
def ReLu(a):
    return np.piecewise(a, [a <= 0.0, a > 0.0], [0, lambda x: x])
```

```python
x = np.linspace(10,-10,100)
print(x)
```

```python
plt.plot(x, sigmoid(x))
```

```python
plt.plot(x, step(x))
```

```python
plt.plot(x, ReLu(x))
```
